# Visualizations of the training process

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Model 3: With pre-training

```python
train_loss = [
    4.618,  # 1
    3.840,
    3.583,  # 3
    3.872,
    3.617,  # 5
    3.487,
    3.393,
    3.320,
    3.260,  # 9
]
val_loss = [
    3.739,  # 1
    3.350,
    3.241,  # 3
    3.562,
    3.494,  # 5
    3.475,
    3.462,
    3.452,
    3.466,  # 9
]
```

```python
plt.figure(figsize=(10,3))

# Declare the figure's contents
yticks = range(1,len(train_loss)+1)
plt.plot(yticks, train_loss, label='Training Loss', color='blue')
plt.axhline(train_loss[-1], color='blue', linestyle='dotted')
plt.plot(yticks, val_loss, label='Validation Loss', color='green')
plt.axhline(val_loss[-2], color='green', linestyle='dotted')
plt.axvline(len(val_loss)-1, color='red')

# Adjust how the figure is displayed
plt.legend()
plt.title("Loss (with fine-tuning) - lower is better")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.gca().xaxis.grid(True)
plt.xlim(left=1, right=len(train_loss))
plt.tight_layout()
plt.savefig('notebook_images/train_loss_with')
plt.show()
```

```python
gleu = [
    9.34,  # 1
    12.06,
    13.38,  # 3
    11.93,
    12.82,  # 5
    11.95,
    13.28,
    13.27,
    13.95,  # 9
]
```

```python
plt.figure(figsize=(10,3))

# Declare the figure's contents
yticks = range(1,len(gleu)+1)
plt.plot(yticks, gleu, label='Validation GLEU', color='green')
plt.axhline(gleu[-2], color='green', linestyle='dotted')
plt.axvline(len(gleu)-1, color='red')

# Adjust how the figure is displayed
plt.legend()
plt.title("GLEU (with fine-tuning) - higher is better")
plt.xlabel("epoch")
plt.ylabel("GLEU")
plt.gca().xaxis.grid(True)
plt.xlim(left=1, right=len(gleu))
plt.tight_layout()
plt.savefig('notebook_images/gleu_with')
plt.show()
```
